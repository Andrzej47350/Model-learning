```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm


df = pd.read_csv('/content/sample_music_data.csv')
data = df[['energy', 'year', 'duration_ms', 'popularity']]

correlation_matrix = data.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

X = data[['energy']]
y = data['popularity']

X = sm.add_constant(X)

ols_model = sm.OLS(y, X).fit()

print(ols_model.summary())

predictions = ols_model.predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X['energy'], y, alpha=0.6, edgecolor='k', label='Data Points')
plt.plot(X['energy'], predictions, color='red', linewidth=2, label='Regression Line')
plt.title("Simple Linear Regression: Popularity vs popularity")
plt.xlabel("energy")
plt.ylabel("popularity")
plt.legend()
plt.show()
```
